```python
import geoviews as gv
gv.extension('bokeh')
from shapely import wkt
from shapely.geometry import Point
```

```python
# Loading test file using xsar
path = xsar.get_test_file('S1A_IW_GRDH_1SDV_20210909T130650_20210909T130715_039605_04AE83_Z010.SAFE')
```

## Using "tiling" module

## Tiling GRD tiles size (17600*17600) resolution = '400m' 

```python
# Centering option desactivated
ds, tiles = grdtiler.tiling_prod(
    path=path, 
    tile_size=17600,
    resolution='400m', 
    detrend=False,
    noverlap=0,
    centering=False,
    side='right',
    save=False,
    save_dir='.',
    to_keep_var=None
)
tiles
```

### Display dataset and tiles footprints

```python
gv.tile_sources.OSM*gv.Polygons(ds.main_footprint)*gv.Polygons(wkt.loads(fp) for fp in tiles.tile_footprint.values)
```

## Tiling GRD tiles size (100*100) resolution = '1000m' option: centering 'right'

```python
# Centering option activated
ds, tiles = grdtiler.tiling_prod(
    path=path, 
    tile_size=17600,
    resolution='400m', 
    detrend=False,
    noverlap=0,
    centering=True,
    side='right',
    save=False,
    save_dir='.',
    to_keep_var=None
)
tiles
```

### Display dataset and tiles footprints

```python
gv.tile_sources.OSM*gv.Polygons(ds.main_footprint)*gv.Polygons(wkt.loads(fp) for fp in tiles.tile_footprint.values)
```

## Tiling GRD tiles size (17600*17600) resolution = '400m' option: centering 'right' noverlap = 20

```python
# Centering option activated with noverlap=20
ds, tiles = grdtiler.tiling_prod(
    path=path, 
    tile_size=17600,
    resolution='400m', 
    detrend=False,
    noverlap=20,
    centering=True,
    side='right',
    save=False,
    save_dir='.',
    to_keep_var=None
)
tiles
```

### Display dataset and tiles footprints

```python
gv.tile_sources.OSM*gv.Polygons(ds.main_footprint)*gv.Polygons(wkt.loads(fp) for fp in tiles.tile_footprint.values).opts(alpha=0.5)
```

## Tiling GRD tiles size (17600*17600) resolution = '400m' around points

```python
point_list = [Point(-108.5, 21.5), Point(-108, 21), Point(-107.5, 21)]
```

```python
ds, tiles = grdtiler.tiling_by_point(
    path=path, 
    posting_loc=point_list,
    tile_size=17600,
    resolution='400m',
    detrend=True,
    save=False,
    save_dir='.',
    to_keep_var=None
)
tiles
```

```python
gv.tile_sources.OSM*gv.Polygons(ds.main_footprint)*gv.Polygons(wkt.loads(fp) for fp in tiles.tile_footprint.values)
```
