# GRDTILER simple exanple

```python
import grdtiler
import xsar
import geoviews as gv
```

```python
# Loading test file using xsar
path = xsar.get_test_file('S1A_IW_GRDH_1SDV_20210909T130650_20210909T130715_039605_04AE83_Z010.SAFE')
```

## Using "tiling" module

## Tiling GRD tiles size (100*100) resolution = '1000m' 

```python
# Centering option desactivated
ds, tiles = grdtiler.tiling_prod(path=path, nperseg=100, resolution='1000m', tiling_mod='tiling', centering=False)
tiles
```

### Geting tiles footprints

```python
# For polarisation VV
tiles_footprint = grdtiler.get_tiles_footprint(tiles, polarization='VV')
tiles_footprint
```

### Display dataset and tiles footprints

```python
plot = gv.tile_sources.OSM*gv.Polygons(ds.footprint)*gv.Polygons(footprint for footprint in tiles_footprint)
plot = plot.opts(title = 'IW_GRD_(resol:1000m, tile:100*100)', width=800, height=800)
plot
```

### Display dataset and tiles on world map

```python
grdtiler.plot_cartopy_data(ds=ds, tiles=tiles, polarization='VV', file_name='map', tiles_footprint=tiles_footprint)
```

## Tiling GRD tiles size (100*100) resolution = '1000m' option: centering 'right'

```python
# Centering option activated
ds, centered_tiles = grdtiler.tiling_prod(path=path, nperseg=100, resolution='1000m', tiling_mod='tiling', centering=True, side='right')
centered_tiles
```

### Display dataset and tiles footprints

```python
# For polarisation VV
centered_tiles_footprint = grdtiler.get_tiles_footprint(centered_tiles, polarization='VV')
centered_tiles_footprint
```

```python
plot = gv.tile_sources.OSM*gv.Polygons(ds.footprint)*gv.Polygons(footprint for footprint in centered_tiles_footprint)
plot = plot.opts(title = 'IW_GRD_(resol:1000m, tile:100*100, centered)', width=800, height=800)
plot
```

### Display dataset and tiles on world map

```python
grdtiler.plot_cartopy_data(ds=ds, tiles=centered_tiles, polarization='VV', file_name='map', tiles_footprint=centered_tiles_footprint)
```

## Tiling GRD tiles size (100*100) resolution = '1000m' option: centering 'right' noverlap = 50

```python
# Centering option activated with noverlap=50
ds, centered_tiles_noverlap = grdtiler.tiling_prod(path=path, nperseg=100, resolution='1000m', tiling_mod='tiling', centering=True, side='right', noverlap=50)
centered_tiles_noverlap
```

### Display dataset and tiles footprints

```python
# For polarisation VV
centered_tiles_noverlap_footprint = grdtiler.get_tiles_footprint(centered_tiles_noverlap, polarization='VV')
centered_tiles_noverlap_footprint
```

### Display dataset and tiles on world map

```python
grdtiler.plot_cartopy_data(ds=ds, tiles=centered_tiles_noverlap, polarization='VV', file_name='map', tiles_footprint=centered_tiles_noverlap_footprint)
```
